#COLUMNA BIRTH RATE

```python
#BUSQUEDA DE DATOS PERDIDOS DE LA COLUMNA 'BIRTH RATE'

df_world_2023_clean[df_world_2023_clean['birth_rate'].isnull()]
```

```python
#REEMPLAZO DE LOS DATOS POR LA MEDIA DE LA COLUMNA 'BIRTH RATE'.

df_world_2023_clean['birth_rate'] = df_world_2023_clean['birth_rate'].fillna(df_world_2023_clean['birth_rate'].mean())
```

```python
#REVISIÓN RESULTADO DEL REMPLAZO DE LA COLUMNA 'BIRTH RATE'

df_world_2023_clean[df_world_2023_clean['birth_rate'].isnull()]
```

```python
#CÁLCULO DEL RANGO INTERCUARTÍLICO PARA CUALQUIER COLUMNA.

def calcular_limites(df, columna):
    Q1 = df[columna].quantile(0.25)
    Q3 = df[columna].quantile(0.75)

    # Calcular el rango intercuartílico (IQR)
    IQR = Q3 - Q1

    # Calcular los límites
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    return limite_inferior, limite_superior

# Llamar a la función para calcular los límites de 'infant_mortality'
limite_inferior, limite_superior = calcular_limites(df_world_2023_clean, 'birth_rate')

# Imprimir los resultados
print("Límite inferior:", limite_inferior)
print("Límite superior:", limite_superior)
```

```python
#CÁLCULO DE LOS PERCENTILES.

df_world_2023_clean['birth_rate'].min(), df_world_2023_clean['birth_rate'].max()
```

```python
#ELIMINACIÓN DE OUTLIERS O DATOS ATÍPICOS EN LA COLUMNA 'BIRTH RATE'.

df_world_2023_clean['birth_rate'] = np.where(df_world['birth_rate'] > limite_superior, np.nan, df_world_2023_clean['birth_rate'])
```
